on traite les dépendances entre l'intégralité des tokens de la meme chanson

```python
import spacy
from spacy import displacy
import pandas as pd

# Charger le modèle spaCy pour l'anglais
nlp = spacy.load('en_core_web_sm')

# Charger votre base de données
df = pd.read_csv('songs _data.csv')

# Sélectionner une chanson spécifique
index_chanson = 450
chanson_texte = df.loc[index_chanson, 'Lyrics']

# Analyser le texte avec spaCy
doc = nlp(chanson_texte)

#visualisation de l'arbre de dépendances
displacy.render(doc, style='dep', jupyter=True, options={'distance': 90})
```

```python
import spacy
from spacy import displacy
import pandas as pd

# Charger le modèle spaCy pour l'anglais
nlp = spacy.load('en_core_web_sm')

# Charger votre base de données
df = pd.read_csv('songs _data.csv')

# Sélectionner une chanson spécifique
index_chanson = 450
chanson_texte_complet = df.loc[index_chanson, 'Lyrics']

# Sélectionner la partie de la chanson que vous souhaitez analyser
debut_texte = 0
fin_texte = 100
chanson_texte_partiel = chanson_texte_complet[debut_texte:fin_texte]

# Analyser le texte avec spaCy
doc = nlp(chanson_texte_partiel)

#visualisation de l'arbre de dépendances
displacy.render(doc, style='dep', jupyter=True, options={'distance': 90})
```
