```python
import csv
import nltk
from langdetect import detect
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Fonction pour appliquer POS à une liste de mots
def pos_tagging(words):
    return nltk.pos_tag(words)

# Charger le fichier CSV contenant les paroles de musique
input_file = 'data_cleaned.csv'  # Mettez le chemin approprié vers votre fichier CSV

# Dictionnaires pour stocker les paroles par langue
lyrics_by_language = {'darija': [], 'francais': [], 'anglais': []}

# Ouvrir le fichier CSV en lecture
with open(input_file, 'r', encoding='utf-8') as file:
    reader = csv.reader(file)

    # Parcourir les lignes du fichier CSV
    for row in reader:
        if len(row) >= 1:  # Assurez-vous que la ligne contient au moins une colonne de texte
            lyric = row[0]  # Assurez-vous que la colonne correspondant aux paroles est correcte

            # Essayez de détecter la langue
            try:
                lang = detect(lyric)
            except:
                lang = 'inconnu'

            # Tokenisation des mots dans les paroles
            words = nltk.word_tokenize(lyric)

            # Stocker les paroles dans le dictionnaire approprié selon la langue détectée
            if lang == 'ar':
                lyrics_by_language['darija'].extend(words)
            elif lang == 'fr':
                lyrics_by_language['francais'].extend(words)
            elif lang == 'en':
                lyrics_by_language['anglais'].extend(words)

# Appliquer POS à chaque liste de mots de chaque langue
pos_tags_by_language = {}
for lang, words in lyrics_by_language.items():
    pos_tags_by_language[lang] = pos_tagging(words)

# Enregistrer les résultats POS par langue dans des fichiers CSV séparés
for lang, pos_tags in pos_tags_by_language.items():
    output_file = f'pos_results_{lang}.csv'

    with open(output_file, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)

        # Écriture des en-têtes pour chaque colonne
        writer.writerow(['Mot', 'POS'])

        # Écrire les résultats POS dans le fichier CSV
        for word, pos in pos_tags:
            writer.writerow([word, pos])
```

```python
import pandas as pd
import stanza
from langdetect import detect

# Téléchargement des modèles Stanza pour différentes langues
stanza.download('fr')
stanza.download('en')

# Initialisation des modèles pour le français et l'anglais
nlp_fr = stanza.Pipeline('fr', processors='tokenize,pos,lemma,depparse')
nlp_en = stanza.Pipeline('en', processors='tokenize,pos,lemma,depparse')

# Charger le fichier CSV contenant les paroles de chansons
file_path = 'data_cleaned.csv'

# Lire le fichier CSV dans un DataFrame Pandas
data = pd.read_csv(file_path)

# Dictionnaires pour stocker les paroles par langue
lyrics_by_language = {'darija': [], 'francais': [], 'anglais': [], 'inconnu': []}

# Analyse des langues et séparation des paroles par langue
for index, row in data.iterrows():
    lyric = row['Lyrics']  # Supposons que la colonne contenant les paroles est nommée 'Lyrics'
    try:
        language = detect(lyric)
    except:
        language = 'inconnu'

    if language == 'ar':
        lyrics_by_language['darija'].append(lyric)
    elif language == 'fr':
        lyrics_by_language['francais'].append(lyric)
    elif language == 'en':
        lyrics_by_language['anglais'].append(lyric)
    else:
        lyrics_by_language['inconnu'].append(lyric)

# Fonction pour effectuer l'analyse des dépendances syntaxiques sur les paroles
def analyze_dependencies(lyrics, nlp, lang):
    all_results = []
    for song in lyrics:
        doc = nlp(song)
        for sent in doc.sentences:
            for word in sent.words:
                head_word = sent.words[word.head - 1].text if word.head > 0 else "ROOT"
                result = [word.text, word.pos, word.lemma, head_word, word.deprel]
                all_results.append(result)
    # Écriture des résultats dans un fichier CSV
    df = pd.DataFrame(all_results, columns=['Word', 'POS', 'Lemma', 'Head', 'Dependency_Relation'])
    df.to_csv(f'pos_results_{lang}.csv', index=False)

# Appliquer l'analyse des dépendances syntaxiques sur les paroles pour chaque langue détectée
for lang, song_lyrics in lyrics_by_language.items():
    if lang != 'inconnu':
        if lang == 'darija':
            analyze_dependencies(song_lyrics, nlp_fr, lang)
        elif lang == 'francais':
            analyze_dependencies(song_lyrics, nlp_fr, lang)
        elif lang == 'anglais':
            analyze_dependencies(song_lyrics, nlp_en, lang)
```
