```python
import re

def keep_words(input_string):
    # Remove '\ax0' or '\0ax' from the input string
    cleaned_string = input_string.replace('\xa0', '').replace('\n', '')
    return cleaned_string
```

```python
North_African_Songs = {}

for j in range(len(List_url)):
  lyrics = ''
  page = requests.get(List_url[j])
  soup = BeautifulSoup(page.text, 'html')
  for i in range(len(soup.find_all('p'))-2):
    extract_p = soup.find_all('p')[i]
    if extract_p.text != '\n' and extract_p.text != '\xa0':
      lyrics += ' ' + keep_words(extract_p.text)
  North_African_Songs[j] = {"lyrics": lyrics, "dialect": 1}
```

```python
import csv


# Specify the CSV file path
csv_file_path = "songs.csv"

# Specify the field names (column names) for the CSV
field_names = ["lyrics", "dialect"]

# Open the CSV file in write mode
with open(csv_file_path, mode="w", newline="", encoding="utf-8") as csv_file:
    # Create a CSV writer
    csv_writer = csv.DictWriter(csv_file, fieldnames=field_names)

    # Write the header to the CSV file
    csv_writer.writeheader()

    # Write the data from the dictionary to the CSV file
    for song_data in North_African_Songs.values():
        csv_writer.writerow(song_data)

print(f"The CSV file '{csv_file_path}' has been created.")
```
